```python
# Step 1: Libraries
import numpy as np
import pandas as pd
import cv2, os, random
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import ConvLSTM2D, BatchNormalization, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# -----------------------------
# Step 2: Load Irradiance CSV
# -----------------------------
path = "/content/drive/MyDrive/2019_09_07"   # <-- change to your dataset path
df = pd.read_csv(f"{path}/pyranometer.csv", header=None)
df.columns = ["timestamp", "irradiance"]

df["timestamp"] = pd.to_datetime(df["timestamp"], format="UTC-7_%Y_%m_%d-%H_%M_%S_%f", errors="coerce")
df = df.set_index("timestamp")

# Keep only daytime
df = df[df["irradiance"] > 0]
# Resample to 5-min average
df = df.resample("5min").mean().dropna() # Changed '5T' to '5min'
print("After resampling:", df.shape)

# -----------------------------
# Step 3: Image Loader
# -----------------------------
def load_image(path, size=(64,64)):
    img = cv2.imread(path)
    if img is None:
        return None
    img = cv2.resize(img, size)
    return img / 255.0

# Using the improved find_closest_image function defined elsewhere
# def find_closest_image(timestamp, image_folder):
#     fname = timestamp.strftime("UTC-7_%Y_%m_%d-%H_%M_%S_%f.jpg")
#     img_path = os.path.join(image_folder, fname)
#     if os.path.exists(img_path):
#         return img_path
#     return None

# -----------------------------
# Step 4: Build sequences
# -----------------------------
X, y = [], []
sequence_len = 6    # 6 images (5-min apart)
forecast_horizon = 6   # predict 30 min later
image_folder = f"{path}/images/"

timestamps = df.index
for i in range(len(timestamps) - sequence_len - forecast_horizon):
    seq_imgs = []
    for j in range(sequence_len):
        ts = timestamps[i+j]
        # Use the improved find_closest_image function
        img_path = find_closest_image(ts, image_folder)
        if img_path:
            img = load_image(img_path)
            if img is not None:
                seq_imgs.append(img)
    # Only append sequence if all images were found
    if len(seq_imgs) == sequence_len:
        X.append(seq_imgs)
        y.append(df.iloc[i+sequence_len+forecast_horizon]["irradiance"])

X = np.array(X)
y = np.array(y)

print("Total sequences:", len(X))
print("X shape:", X.shape, " y shape:", y.shape)

# -----------------------------
# Step 5: Train-test split
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

# -----------------------------
# Step 6: ConvLSTM model
# -----------------------------
def build_model():
    model = Sequential([
        ConvLSTM2D(filters=32, kernel_size=(3,3), activation='relu',
                   input_shape=(sequence_len,64,64,3), return_sequences=True),
        BatchNormalization(),
        ConvLSTM2D(filters=32, kernel_size=(3,3), activation='relu', return_sequences=False),
        BatchNormalization(),
        Flatten(),
        Dense(64, activation='relu'),
        Dense(1)  # regression output
    ])
    model.compile(optimizer=Adam(1e-4), loss="mse", metrics=["mae"])
    return model

# -----------------------------
# Step 7: Train with retry if RÂ² â‰¤ 0
# -----------------------------
attempt = 1
r2 = -1
while r2 <= 0:
    print(f"\nðŸš€ Training attempt {attempt} ...")
    model = build_model()
    early_stop = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)
    history = model.fit(
        X_train, y_train,
        validation_split=0.2,
        epochs=20,
        batch_size=16,
        callbacks=[early_stop],
        verbose=1
    )
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    attempt += 1

# -----------------------------
# Step 8: Final Evaluation
# -----------------------------
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print("\nâœ… Final Results")
print(f"RÂ² Score: {r2:.4f} (Positive âœ…)")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")

# Scatter Plot
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Actual Irradiance")
plt.ylabel("Predicted Irradiance")
plt.title(f"Actual vs Predicted (RÂ²={r2:.3f})")
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()], "r--")
plt.show()

# Time-Series Plot
plt.figure(figsize=(12,5))
plt.plot(y_test[:200], label="Actual")
plt.plot(y_pred[:200], label="Predicted")
plt.xlabel("Samples")
plt.ylabel("Irradiance")
plt.title("Actual vs Predicted (Time Series)")
plt.legend()
plt.show()
```
